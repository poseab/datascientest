### Data frame structure

* **text_ocr** - the exctracted text before OCR process

* **text** - normalized text_ocr (lowercase, regex tokenizer, delete stop words)
* **text_fr** - french translated text_ocr with minimum normalization (delete numbers, delete dates)
* **text_en** - english translated text_ocr with minimum normalization (delete numbers, delete dates)
* **text_fr_norm** - normalized text_fr (lowercase, regex tokenizer, delete stop words, delete punctuations)
* **text_en_norm** - normalized text_en (lowercase, regex tokenizer, delete stop words, delete punctuations)

**Targets**

* **target** - target
* **target_min** - some regrupments have been done to simplify the target

* target_en - english labels (dont use it)
* target_fr - french labels (dont use it)
