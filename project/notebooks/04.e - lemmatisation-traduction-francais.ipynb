```python
import pandas as pd
df=pd.read_csv('df_fr_new.csv')
df.head()
df=df.dropna()
```

```python
from transformers import pipeline
en_fr_trnslator=pipeline('translation_en_to_fr')

def tokn_phrase_traduction(text):
    
    list_phrase=str(text).lower().replace(',','___').replace('.','___').split('___')
    
    return ' '.join([en_fr_trnslator(i)[0]['translation_text'] for i in list_phrase ])
```

```python
df['text_image_fr_new']=[tokn_phrase_traduction(txt) for txt in df.text_image]
df.to_csv('df_en_to_fr.csv')




```
